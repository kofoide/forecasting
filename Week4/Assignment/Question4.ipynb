```r
# read in the csv file
ds <- read.csv("../../pigs.csv", sep=",", header=TRUE)

# use zoo to convert the date "YYYY-MM" to a date
pigzoo <- read.zoo(ds, sep=",", FUN=as.yearmon)

# us ts to convert the zoo object to ts object using the correct intervals
pigs = ts(pigzoo, frequency=12, start=c(1980,1), end=c(1995, 8))
```

## 4a)

* Trend - for the most part increases, except for a big down swing from 1985 to 1988 (not due to swine flu)
* Seasonality - definitely
* Cycle - there could be a cycle in here, would be able to see if I had more data

```r
plot(pigs)
```

## 4b)

Holt-Winters is best since this data has both Trend & Seasonality

Damped-Trend Multiplicative-Seasonality

```r
hw_mult_damp = hw(pigs, seasonal="multiplicative", damped=TRUE, h=24)

plot(pigs, main="Pigs with Forecast from HW Mult & Damp", xlim=c(1979, 1998))

# Plot the fitted values to original
lines(fitted(hw_mult_damp), col="blue")
# Plot the forecasted values
lines(hw_mult_damp$mean, col="blue")
```

```r
summary(hw_mult_damp)
```

```r
hw_mult = hw(pigs, seasonal="multiplicative", damped=FALSE, h=24)

plot(pigs, main="Pigs with Forecast from Holt-Winters Multiplicative Season", xlim=c(1979, 1998))
# Plot the fitted values to original
lines(fitted(hw_mult), col="red")
# Plot the forecasted values
lines(hw_mult$mean, col="red")

# Plot the fitted values to original
lines(fitted(hw_mult_damp), col="blue")
# Plot the forecasted values
lines(hw_mult_damp$mean, col="blue")
```

Now I think I need to put an emphasis on the more recent observations, bigger alpha

```r
hw_mult_alpha = hw(pigs, seasonal="multiplicative", damped=FALSE, h=24, alpha=0.8)

plot(pigs, main="Pigs with Forecast from Holt-Winters Multiplicative Season", xlim=c(1979, 1998))
# Plot the fitted values to original
lines(fitted(hw_mult_alpha), col="red")
# Plot the forecasted values
lines(hw_mult_alpha$mean, col="red")
```

```r
summary(hw_mult_alpha)
```

```r
hw_mult_alpha_exp = hw(pigs, seasonal="multiplicative", damped=FALSE, exponential=TRUE, h=24, alpha=0.5)

plot(pigs, main="Pigs with Forecast from Holt-Winters Multiplicative Season", xlim=c(1979, 1998))
# Plot the fitted values to original
lines(fitted(hw_mult_alpha), col="red")
# Plot the forecasted values
lines(hw_mult_alpha$mean, col="red")
```
